```python
files = ["new_generating_units", "planned_generating_additions", "planned_generating_retirements", "retired_generating_units"]
```

```python
def extract_pv_generating_units(filename):
    #read in file
    df = pd.read_csv(f"./clean/{filename}.csv")
    
    #print duplicate rows and number null values
    df.drop_duplicates(inplace=True)
    print("dupes", df.duplicated().sum())
    print("nulls", df.isnull().sum())
    
    #return PV rows
    new_filename = f"./clean/cleaned_{filename}.csv"
    df = df[df["Technology"]=="Solor Photovoltaic"].sort_values("Year")
    df.to_csv(new_filename)
```

```python
for file in files:
    extract_pv_generating_units(file)
```
